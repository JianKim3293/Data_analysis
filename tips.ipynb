```python
# numerical calculation & data frames
import numpy as np
import pandas as pd

# visualization
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn.objects as so
import plotly.express as px

# statistics
import statsmodels.api as sm
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
houses_data = sm.datasets.get_rdataset("SaratogaHouses", "mosaicData")

houses = houses_data.data # data
print(houses_data.__doc__) # documents
```

**Scatter Plot**

```python
# 전체 식사비용(total_bill)에 따른 tip을 얼마나 지불했는지
(
    so.Plot(tips, x="total_bill", y="tip") # empty plot을 생성하고, x, y축에 mapping할 mpg 데이터의 변수를 지정
    .add(so.Dot()) # layer를 추가하여, points들을 Dot이라는 mark object를 써서 표현
)
```

```python
# 컬럼 time 에 따른 데이터 구분
(
    so.Plot(tips, x="total_bill", y="tip", color="time")
    .add(so.Dot())
)
# ax = sns.scatterplot(x='total_bill', y='tip', hue='time',data=tips)
```

```python
# 요일별 시간에 따른 마커 변경
(
    so.Plot(tips, x="total_bill", y="tip", marker="time")
    .add(so.Dot())
)
# ax = sns.scatterplot(x='total_bill', y='tip', hue='day', style='time',data=tips)
```

```python
# Alpha
(
    so.Plot(tips, x="total_bill", y="tip", alpha="time")
    .add(so.Dot())
)
```

```python
(
    so.Plot(tips, x="total_bill", y="tip", color="time", marker="smoker")
    .add(so.Dot())
)
```

```python
sns.catplot(
    data=tips, kind="swarm",
    x="total_bill", y="tip", hue="smoker", col="sex",
    aspect=.6
)
plt.show()
```

```python

p = so.Plot(tips, x="total_bill", y="tip", linestyle="day", color="day")
p.add(so.Line())
```

```python
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("요일 별 전체 팁의 Strip Plot")
plt.show()
```

```python
sns.histplot(tips, x="tip", y="total_bill")
plt.show()
```

**Faceting**

```python
t = (
    so.Plot(tips, x="tip", y="total_bill")
    .facet(col="day", row="sex")
    .add(so.Dot(alpha=.5))
)
t
```

**Pairing**

```python
(
    so.Plot(tips, y="total_bill", color="time") # y축은 공유
    .pair(x=["tip", "sex"]) # x축에 다른 변수를 mapping
    .add(so.Dots()) # .Dots()! overploting에 유리. .Dot(alpha=.)로도 비슷
)
```

**Facet & pair 동시**

```python
(
    so.Plot(tips, y="tip", color="sex")
    .pair(x=["total_bill", "smoker"])
    .facet(row="day")
    .add(so.Dots())
)
```

**Fitted lines**

```python
(
    so.Plot(tips, x="total_bill", y="tip")
    .add(so.Dot(), color="sex")
    .add(so.Line(), so.PolyFit(5), group="day") # color가 아닌 group으로 grouping
)
# 다항함수 fit의 특징 및 주의점
```

# Count plot

**Bar plot**

```python
sns.countplot(x="day", data=tips)
plt.title("요일별 팁을 준 횟수")
plt.show()
```

```python
(
    so.Plot(tips, x="day")
    .add(so.Bars(), so.Hist()) # Histogram; x값을 bins으로 나누어 count를 계산!
    # .Bars()는 .Bar()에 비해 연속변수에 더 적합: 얇은 경계선으로 나란히 붙혀서 그려짐
)
```

```python
# Density plot: 넓이가 1이 되도록
(
    so.Plot(tips, x="day")
    .add(so.Area(), so.KDE()) # Density plot
)
```

```python
# Density plot: 넓이가 1이 되도록
(
    so.Plot(tips, x="day")
    .add(so.Line(color="orange"), so.KDE(bw_adjust=.2)) # Density bandwidth: binwidth에 대응
    .add(so.Bars(alpha=.3), so.Hist("density")) # stat="density"
)
```

**boxplot**
명령은 박스-휘스커 플롯(Box-Whisker Plot) 혹은 간단히 박스 플롯이라 부르는 차트를 그려준다. 박스 플롯은 박스와 박스 바깥의 선(whisker)으로 이루어진다.

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 전체 팁의 Box Plot")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("요일 별, 성별 전체 팁의 Histogram")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("요일 별, 성별 전체 팁의 Box Plot")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")
plt.show()
```

```python
(
    so.Plot(tips, x="day", y="total_bill")
    .add(so.Dots(color=".5"), so.Jitter()) # so.Jitter(): 흐트려뜨려 그리기
    .add(so.Dot(pointsize=8), so.Agg("median")) # .Agg(): aggregation, default는 mean
    .add(so.Range(), so.Est("median", errorbar=("pi", 50))) # .Range(): 기본 min/max range, .Est(): estimator
)
```

```python
(
    so.Plot(tips, x="day", y="total_bill", color="smoker")
    .add(so.Dots(), so.Jitter(), so.Dodge())
    .add(so.Dot(pointsize=5), so.Agg("median"), so.Dodge())
    .add(so.Range(), so.Est("median", errorbar=("pi", 50)), so.Dodge())
)
```

```python
(
    so.Plot(tips, x="day", y="total_bill", color="sex")
    .add(so.Dots(), so.Jitter(), so.Dodge())
    .add(so.Dot(pointsize=5), so.Agg("median"), so.Dodge())
    .add(so.Range(), so.Est("median", errorbar=("pi", 50)), so.Dodge())
)
```

**Histogram**

```python
(
    so.Plot(tips, x="total_bill", color="day")
    .add(so.Bar(), so.Hist(binwidth=200, common_bins=False))  # bins을 공유하지 않도록
)
# Hist(): 다양한 parameter가 있음...
```

```python
(
    so.Plot(tips, x="day", color="tip")
    .add(so.Bar(), so.Hist(binwidth=200, common_bins=True))  
)
# Hist(): 다양한 parameter가 있음...
```

```python
(
    so.Plot(tips, x="sex", color="tip")
    .add(so.Bar(), so.Hist(binwidth=200, common_bins=False))  # bins을 공유하지 않도록
)
# Hist(): 다양한 parameter가 있음...
```

**Frequency polygon**

```python
(
    so.Plot(tips, x="day", color="time")
    .add(so.Line(marker="."), so.Hist(binwidth=200)) # Line에 maker "."을 표시
)
```

```python
(
    so.Plot(tips, x="day", color="smoker")
    .add(so.Line(marker="."), so.Hist(binwidth=200, stat="proportion",  common_norm=False)) # Line에 maker "."을 표시
)
```

**Density plot**

```python
(
    so.Plot(tips, x="total_bill", color="day")
    .add(so.Area(), so.KDE(common_norm=False)) # Density plot, species별로 넓이가 1이 되도록
)
```

```python
p = so.Plot(tips, x="day", color="smoker")
p.add(so.Bar(), so.Count()) # Bar() mark + Count() transformation
```

```python
left = p.add(so.Bar(), so.Count(), so.Dodge()) # 나란히 표시
right = p.add(so.Bar(), so.Count(), so.Stack()) # stacking
```

```python
p.add(
    so.Bar(width=.5), so.Hist("proportion"), # proportion; stat="count"로 하면 앞서 so.Count()와 동일
    so.Stack() # stacking
)
```

```python
p.add(
    so.Bar(width=.5), so.Hist("proportion", common_norm=["x"]), # proportion; 
    so.Stack() # stacking
)
```

```python
p.add(
    so.Bar(width=.8), so.Hist("proportion", common_norm=["x", "col"]), # proportion
    so.Stack(), # stacking
).facet(col="sex") # faceting
```
