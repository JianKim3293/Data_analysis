```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('Popular_baby_names.xml')
root = tree.getroot()
```

```python
for child in root.iter():
    print(child.tag)
```

```python
rows=[]

for i in root.findall('row/row'):
  lst=[]
  id = i.attrib['_id']
  uuid = i.attrib['_uuid']
  address = i.attrib['_address']
   
  first_name = i.findtext('first_name')
  last_name = i.findtext('last_name')
  city = i.findtext('city')    
    
  lst.append(id)
  lst.append(uuid)
  lst.append(address)
  lst.append(first_name)
  lst.append(last_name)
  lst.append(city)
  rows.append(lst)
    
    
rows
```

```python
import pandas as pd
df= pd.DataFrame(rows ,columns=['ID', 'uuid', 'Address', 'First_name', 'Last_name', 'City'])
df
```

```python
missing_df = df.isnull()
for col in missing_df.columns:
  missing_count = missing_df[col].value_counts()
  #print(missing_count)
  try:
    print(col,': ', missing_count[True]) # true가 있는것의 개수를 출력 
  except:
    print(col,": ",0)
```

```python
cnt=0
for value in df['City']:
  if value is None:
    cnt +=1

print(cnt,'개의 None 값이 있습니다.')
```

누락데이터의 개수가 8개 뿐이므로 city의 값을 대체할 자료가 없기 때문에 데이터 삭제로 처리

```python
df_cleaned = df.dropna(subset=['City'], how='any')
len(df_cleaned)
```
